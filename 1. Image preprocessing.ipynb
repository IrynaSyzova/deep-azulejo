```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
img = cv2.imread('data/examples/example2.jpg')[...,::-1]
```

```python
plt.imshow(img)
```

```python
blur1 = cv2.medianBlur(img, int(img.shape[0] * 0.001) * 2 + 1)
blur2 = cv2.medianBlur(img, int(img.shape[0] * 0.002) * 2 + 1)
blur3 = cv2.medianBlur(img, int(img.shape[0] * 0.003) * 2 + 1)
blur4 = cv2.medianBlur(img, int(img.shape[0] * 0.004) * 2 + 1)
blur5 = cv2.medianBlur(img, int(img.shape[0] * 0.005) * 2 + 1)

plt.figure(figsize=(16, 9))

titles = [
    'Original Image',
    'MedianBlur={0}'.format(int(img.shape[0] * 0.001) * 2 + 1),
    'MedianBlur={0}'.format(int(img.shape[0] * 0.002) * 2 + 1),
    'MedianBlur={0}'.format(int(img.shape[0] * 0.003) * 2 + 1),
    'MedianBlur={0}'.format(int(img.shape[0] * 0.004) * 2 + 1),
    'MedianBlur={0}'.format(int(img.shape[0] * 0.005) * 2 + 1)]
images = [img, blur1, blur2, blur3, blur4, blur5]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
