```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
img = cv2.imread('dataset/azulejo/000008.jpg')[...,::-1]
plt.imshow(img)
```

Gaussian blur:

```python
kernel_sizes = [int(img.shape[0] * (x + 1) / 100.0) * 2 + 1 for x in range(5)]
print('Kernel sizes: {0}'.format(kernel_sizes))

images = [img] + [
    cv2.GaussianBlur(img, (ks, ks), 0)
    for ks in kernel_sizes
]

titles = ['Original Image'] + [
    'GaussianBlur={0}'.format(ks)
    for ks in kernel_sizes
]

plt.figure(figsize=(16, 9))

for i in range(len(images)):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
